```python
x = np.array(range(0,10))

y1 = np.sqrt(x)
y2 = np.power(x,3)
y3 = np.power(2,x)
y4 = 1/(x+1)

plt.figure(figsize = (8,8))
plt.suptitle('Plot exercise')

plt.subplot(221)
plt.plot(x,y1, color = 'r')
plt.xlabel('x')
plt.ylabel('y1')

plt.subplot(222)
plt.plot(x,y2, color = 'g')
plt.xlabel('x')
plt.ylabel('y2')

plt.subplot(223)
plt.plot(x,y3, color = 'm')
plt.xlabel('x')
plt.ylabel('y3')

plt.subplot(224)
plt.plot(x,y4, color = 'k')
plt.xlabel('x')
plt.ylabel('y4')
```

```python
- 4.Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.
```

```python
x = np.array(range(0,3))

y1 = np.sqrt(x)
y2 = np.power(x,3)
y3 = np.power(2,x)
y4 = 1/(x+1)

plt.plot(x,y1, c='r')
plt.plot(x,y2, c='g')
plt.plot(x,y3, c='m')
plt.plot(x,y4, c='k')

plt.legend(['y1','y2','y3','y4'])
```
