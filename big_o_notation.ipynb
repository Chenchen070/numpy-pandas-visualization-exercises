```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
n = np.array(range(1,4))
chart = plt.figure(figsize = (16,8))
plt.suptitle('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')

y1 = 0*n + 1
plt.plot(n,y1, label = '$\O(1)$', color = 'r')

y2 = np.log(n)
plt.plot(n,y2, label = '$\O(log n)$', color = 'm')

y3 = n
plt.plot(n,y3, label = '$\O(n)$', color = 'g')

y4 = n*np.log(n)
plt.plot(n,y4, label = '$\O(n log n)$', color = 'b')

y5 = np.square(n)
plt.plot(n,y5, label = '$\O(n^2)$', color = 'gray')

y6 = np.power(2,n)
plt.plot(n,y6, label = '$\O(2^n)$', color = 'k')

import math
from scipy.special import factorial
y7 = factorial(n)
plt.plot(n,y6, label = '$\O(n!)$', color = 'y')

y8 = np.power(n,n)
plt.plot(n,y8, label = '$\O(n^n)$', color = 'c')

chart.text(0.4,0.8, 'Chenchen Feng', style='italic',
        bbox={'facecolor': 'red', 'alpha': 0.2, 'pad': 20})

plt.legend(loc = 'upper left', fontsize = 14)
```

```python
import math
import matplotlib.pyplot as plt

# x values

x = range(1, 20)

# O(1)
y1 = [1 for n in x]

# O(log n)
y2 = [math.log(n) for n in x]

# O(n)
y3 = x

# O(n log n)
y4 = [n * math.log(n) for n in x]

# O(n^2)
y5 = [n**2 for n in x]

# O(2^n)
y6 = [2**n for n in x]



# O(n!)
y7 = [math.factorial(n) for n in x]

# O(n^n)
y8 = [n**n for n in x]
```

```python
plt.figure(figsize = (16, 8))

plt.title('Big O Notation', fontsize = 20)

plt.xlabel('Elements', fontsize=14)
plt.ylabel('Operations', fontsize=14)

# O(1)
plt.plot(x, y1, label='$\mathcal{O}(1)$', color='cyan')

# lightcoral, orange, mediumseagreen, cadetblue, dodgerblue, slateblue, violet, crimson

# O(log n)
plt.plot(x, y2, label='$\mathcal{O}(logn)$', color='lime')

# O(n)
plt.plot(x, y3, label='$\mathcal{O}(n)$', color='green')

# O(n log n)
plt.plot(x, y4, label='$\mathcal{O}(nlogn)$', color='olive')

# O(n^{2})
plt.plot(x, y5, label='$\mathcal{O}(n^{2})$', color='orange')

# O(2^{n})
plt.plot(x, y6, label='$\mathcal{O}(2^{n})$', color='saddlebrown')

# O(n!)
plt.plot(x, y7, label='$\mathcal{O}(n!)$', color='salmon')

# O(n^{n})
plt.plot(x, y8, label='$\mathcal{O}(n^{n})$', color='red')


# Set limits for axis
plt.xlim(1, 11)
plt.ylim(0, 50)

# show legend()
plt.legend(loc='upper right', fontsize=14)

plt.show()
```

```python
# Think Algoritmic efficiency/

# How does the run time scales with input variable?
# Stays same O(1)
# Scales Linearly O(N)
# Scales as square of N (N^2)



# O(1) describes an algorithm that will always execute in the same time regardless
# of the size of the input data set.

# Example: Find first element from a list with n elements.
# We describe this as O(1). 

example_list = [1,2,4,6,7]

print(example_list[0])
```

```python
# O(N) describes an algorithm whose performance will grow linearly and in direct proportion
# to the size of the input data set. O(N) - scales linearly Wrt input. Twice the size of data will take twice as long


n = 5

for i in range(0, n):
    if i % 2 == 0:
        print(i)
```

```python
# %%timeit -r 1 -n 1
# O(N2) represents an algorithm whose performance is directly proportional to the square 
# of the size of the input data set. This is common with algorithms that involve nested iterations
# over the data set
n = 2
# For example 
for i in range(n+1):
    for j in range(n+1):
        print(i+j)
```
